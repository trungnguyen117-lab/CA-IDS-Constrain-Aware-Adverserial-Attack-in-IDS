```python
import pandas as pd

# Đọc dữ liệu
df2 = pd.read_csv("./DS/synthetic_data_CTGAN_9600.csv")

# Giả sử cột nhãn tên là 'Label' (bạn thay lại đúng tên cột nếu khác)
label_col = "Label"

# 1. Lấy toàn bộ các mẫu có nhãn 7, 8, 9, 10
selected_labels = [7, 8, 9, 10]
df_selected = df2[df2[label_col].isin(selected_labels)]

# 2. Số lượng đã chọn
count_selected = len(df_selected)

# 3. Lấy thêm từ các nhãn còn lại cho đủ 4800
remaining_needed = 4800 - count_selected
df_remaining = df2[~df2[label_col].isin(selected_labels)]

# Nếu remaining_needed > 0 thì mới cần sample
if remaining_needed > 0:
    df_remaining_sampled = df_remaining.sample(n=remaining_needed, random_state=42)
    final_df = pd.concat([df_selected, df_remaining_sampled], axis=0)
else:
    # Nếu số lượng nhãn 7-10 đã >= 4800 thì chỉ lấy ngẫu nhiên 4800 trong đó
    final_df = df_selected.sample(n=4800, random_state=42)

# Reset lại index
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Lưu ra file nếu cần
final_df.to_csv("./DS/synthetic_data_CTGAN_4800.csv", index=False)

print("Final dataset shape:", final_df.shape)
print(final_df[label_col].value_counts())
```

```python
import pandas as pd

# === Load data ===
df = pd.read_csv("./DS/train_tabdiff.csv")
LABEL = "Label"

# === Tham số ===
target_total = 4800
special_classes = [7, 8, 9, 10]
special_n = 800  # số mẫu mỗi lớp đặc biệt

# === Lấy 800 mẫu cho mỗi lớp 7,8,9,10 ===
dfs_selected = []
for cls in special_classes:
    subset = df[df[LABEL] == cls]
    sampled = subset.sample(n=special_n, random_state=42)
    dfs_selected.append(sampled)

# === Tính số mẫu còn lại ===
used = special_n * len(special_classes)
remaining = target_total - used

# Các lớp còn lại
other_classes = [c for c in df[LABEL].unique() if c not in special_classes]
per_class = remaining // len(other_classes)

for cls in other_classes:
    subset = df[df[LABEL] == cls]
    sampled = subset.sample(n=per_class, random_state=42)
    dfs_selected.append(sampled)

# === Gộp lại ===
final_df = pd.concat(dfs_selected).sample(frac=1, random_state=42).reset_index(drop=True)

df0 = pd.read_csv("./DS/train_shap_66.csv")

merged = pd.concat([df0, final_df], ignore_index=True)

print(merged[LABEL].value_counts())

# === Lưu file ===
merged.to_csv("./DS/train_tabdiff_9600.csv", index=False)
```
